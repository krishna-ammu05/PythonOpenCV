### Adding NOise to the image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import clear_output
```

```python
# Load the image
img = cv2.imread('pictures/cloud.jpg')

# Generate random Gaussian noise
mean = 0
stddev = 180
noise = np.zeros(img.shape, np.uint8)  # Create an empty image (same shape)
cv2.randn(noise, mean, stddev)  # Fill with Gaussian noise

# Add noise to image
noisy_img = cv2.add(img, noise)

# Save noisy image
cv2.imwrite('pictures/noisy_cloud.jpg', noisy_img)

# Read and display the noisy image
imshow = cv2.imread('pictures/noisy_cloud.jpg')
plt.imshow(cv2.cvtColor(imshow, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

#### Resizing and Rescaling 

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt   #  fixed import

img = cv.imread("pictures/dog.jpg")
cv.imshow("Original image",img)
# Function to rescale frames   works for images,videos  
def rescaleFrame(frame, scale=0.75):
    width = int(frame.shape[1] * scale)   #  correct width
    height = int(frame.shape[0] * scale)  #  correct height
    dimensions = (width, height)
    return cv.resize(frame, dimensions, interpolation=cv.INTER_AREA)

resized_img = rescaleFrame(img)
cv.imshow("resized image",resized_img)

# Read video
capture = cv.VideoCapture('videos/ammujay.mp4')
while True:
    isTrue, frame = capture.read()
    if not isTrue:   #  stop when video ends
        break
    frame_resized = rescaleFrame(frame)
    
    cv.imshow('Video', frame)
    cv.imshow('Video Resized', frame_resized)
    
    if cv.waitKey(20) & 0xFF == ord('d'):
        break

def changeRes(width,height):#only works for live video
    capture.set(3,width)
    capture.set(4,height)
    
capture.release()   #  release capture
cv.destroyAllWindows()

```
