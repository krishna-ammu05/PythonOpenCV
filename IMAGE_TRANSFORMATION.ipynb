```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('pictures/dog.jpg') 
cv.imshow("Dog",img)

def show(title,image):
    plt.imshow(image)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

#### Resizing

```python
resize = cv.resize(img,(300,200))
show("Resize",resize)
```

#### 2.Roatate(90 clockwise)

```python
rotate = cv.rotate(img, cv.ROTATE_90_CLOCKWISE)
rotate1= cv.rotate(img, cv.ROTATE_180)
rotate2= cv.rotate(img, cv.ROTATE_90_COUNTERCLOCKWISE)
show("Rotate",rotate)
show("Rotate",rotate1)
show("Rotate",rotate2)
```

#### 3.Flip

```python
flip = cv.flip(img, 1)
flip1= cv.flip(img, 0)
show("Flip",flip)
show("Flip",flip1)
```

#### 4.Affine Transformation

```python
src_pts = np.float32([[50,50],[200,50],[50,200]])
dts_pts = np.float32([[10,100],[200,50],[100,250]])
M_affine = cv.getAffineTransform(src_pts,dts_pts)
wrapAffine = cv.warpAffine(img,M_affine,(img.shape[1],img.shape[0]))
show("wrapAffine",wrapAffine)
```

#### 5.Wrap perspective

```python
src_pts = np.float32([[50,50],[450,50],[50,400],[450,400]])
dst_pts = np.float32([[10,100],[400,50],[100,300],[400,300]])
M_perspective = cv.getPerspectiveTransform(src_pts, dst_pts)
warpPerspective = cv.warpPerspective(img, M_perspective, (img.shape[1], img.shape[0]))
show("WarpPerspective", warpPerspective)
```

#### 6. Rotation matrix

```python
(h, w) = img.shape[:2]
center = (w//2, h//2)
M_rotate = cv.getRotationMatrix2D(center, 45, 1.0)
getRotationMatrix2D = cv.warpAffine(img, M_rotate, (w, h))
show("getRotationMatrix2D", getRotationMatrix2D)
```

#### 7.Transpose

```python
transpose = cv.transpose(img)
show("transpose",transpose)
```

#### 8.crop

```python
crop = img[40:300 , 40:600]
show("crop",crop)
```
