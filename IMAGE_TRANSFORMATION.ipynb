```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('pictures/dog.jpg') 
cv.imshow("Dog",img)

def show(title,image):
    plt.imshow(image)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

#### Resizing

```python
resize = cv.resize(img,(300,200))
show("Resize",resize)
```

#### 2.Roatate(90 clockwise)

```python
rotate = cv.rotate(img, cv.ROTATE_90_CLOCKWISE)
rotate1= cv.rotate(img, cv.ROTATE_180)
rotate2= cv.rotate(img, cv.ROTATE_90_COUNTERCLOCKWISE)
show("Rotate",rotate)
show("Rotate",rotate1)
show("Rotate",rotate2)
```

#### 3.Flip

```python
flip = cv.flip(img, 1)
flip1= cv.flip(img, 0)
show("Flip",flip)
show("Flip",flip1)
```
