#### Reading a video

````python
import cv2
from matplotlib import pyplot as plt
from IPython.display import clear_output
import time

cap = cv2.VideoCapture('videos/ammujay.mp4')

if not cap.isOpened():
    print("Error: could not open the video")
else:
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Convert BGR to RGB for correct colors in matplotlib
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Clear previous output so frames dont stack
        clear_output(wait=True)

        # Show the frame inline
        plt.imshow(frame_rgb)
        plt.axis('off')
        plt.show()

        # Delay to control playback speed (similar to waitKey)
        time.sleep(0.025)

cap.release()
````

#### Saving image

````python

# Read an image
img = cv2.imread('pictures/dog.jpg')

# Check if image loaded
if img is None:
    print("Error: Could not read the image")
else:
    # Save the image (PNG format here)
    cv2.imwrite('pictures/output.png', img)
    print("Image saved successfully!")
imshow= cv2.imread("pictures/output.png")
plt.imshow(cv2.cvtColor(imshow, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
````


# cv2.imwrite() in OpenCV
## Save a Rondomly Genertaed Image
### **Purpose**
Saves an image (NumPy array) to a file on your local storage.

---

### **Syntax**
```python
cv2.imwrite(path, image)
```
path  Full path (including filename & extension) where the image will be saved.
image  NumPy array representing the image.


### **Return Value**
True  Image successfully saved.
False  Saving failed (e.g., invalid path or permissions).

````python
img = np.random.randint(255, size=(300,600,3), dtype = np.uint8)
new_img = cv2.imwrite("pictures/image.png",img)
if new_img:
    print("the image is asved succesfully ")
imshow = cv2.imread("pictures/image.png")
plt.imshow(cv2.cvtColor(imshow,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
````
