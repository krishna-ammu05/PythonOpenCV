#### COLOR CHANNELS

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread("pictures/house.jpeg")
cv.imshow("image",img)
```

```python
b, g, r = cv.split(img)
#show function
def show(img,title,cmap_type = None):
    plt.figure()
    plt.imshow(img,cmap = cmap_type)
    plt.title(title)
    plt.axis("off")
```

```python
#showing all the images with their channels
show(b,"Blue channel","gray")
show(g,"Green channel","gray")
show(r,"Red channel","gray")
show(cv.merge([b,g,r]),"Merged Image")
```

```python
blank = np.zeros(img.shape[:2],dtype = 'uint8')
blue = cv.merge([b,blank,blank])
green = cv.merge([blank,g,blank])
red = cv.merge([blank,blank,r])
```

```python
show(blue,"Blue channel","gray")
show(green,"Green channel","gray")
show(red,"Red channel","gray")
```

```python
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread("pictures/house.jpeg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Split channels
b, g, r = cv2.split(img_rgb)

# Create subplots (2 rows x 2 cols)
fig, axs = plt.subplots(3, 2, figsize=(10, 8))

# Blue channel
axs[0, 0].imshow(b, cmap="gray")
axs[0, 0].set_title("Blue Channel")
axs[0, 0].axis("on")
# Green channel
axs[0, 1].imshow(g, cmap="gray")
axs[0, 1].set_title("Green Channel")
axs[0, 1].axis("off")

# Red channel
axs[1, 0].imshow(r, cmap="gray")
axs[1, 0].set_title("Red Channel")
axs[1, 0].axis("off")

# Merged image
axs[1, 1].imshow(cv2.merge([b, g, r]))
axs[1, 1].set_title("Merged Image")
axs[1, 1].axis("off")

# Red channel
axs[2, 0].imshow(r, cmap="gray")
axs[2, 0].set_title("Red Channel")
axs[2, 0].axis("off")

# Show all at once
plt.tight_layout()
plt.show()
```
