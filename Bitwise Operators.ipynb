#### Bitwise Operators

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
blank = np.zeros((400,400),dtype = 'uint8')
rectangle = cv.rectangle(blank.copy(),(30,30),(370,370),255,-1)
circle = cv.circle(blank.copy(),(200,200),200,255,-1)
def show(img, title):
    plt.figure(figsize=(4,4))
    plt.imshow(img, cmap="gray")
    plt.title(title)
    plt.axis("off")
    plt.show()


show(rectangle,"Rectangle")
show(circle,"Circle")

```

#### BItwise and operator--->Intersecting Regions

```python
bitwise_and = cv.bitwise_and(rectangle,circle)
show(bitwise_and,"Bitwise_and")
```

#### BItwise OR operator--->Non intersecting Regions and intersecting

```python
bitwise_or = cv.bitwise_or(rectangle,circle)
show(bitwise_or,"Bitwise_or")
```

#### Bitwise Xor --->non intersecting regions

```python
bitwise_xor = cv.bitwise_xor(rectangle,circle)
show(bitwise_xor,"Bitwise_XOR")
```

#### Bitwise NOT

```python
bitwise_not = cv.bitwise_not(rectangle,circle)
show(bitwise_not,"Bitwise_not")
```
