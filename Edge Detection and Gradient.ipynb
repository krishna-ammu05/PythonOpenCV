#### Edge Dectection and Gradient

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(titles,images,cmap = 'gray'):
    n = len(images)
    plt.figure(figsize = (15,8))
    for i in range(n):
        plt.subplot(2,(n+1)//2 ,i+1)
        plt.imshow(images[i],cmap = cmap)
        plt.title(titles[i])
        plt.axis("off")
    plt.show()
```

```python
img = cv.imread("pictures/dog.jpg", cv.IMREAD_GRAYSCALE)  
if img is None:
    raise FileNotFoundError("Image not found. Please check the path.")
```

#### 1.canny Edge Dectection

```python
canny = cv.Canny(img,100,200)
```

#### 2. Sobel Operator(X and Y gradient)

```python
sobelx = cv.Sobel(img,cv.CV_64F,1,0,ksize=3)
sobely = cv.Sobel(img,cv.CV_64F,0,1,ksize=3)
sobel = cv.magnitude(sobelx,sobely)
```

#### 3.Laplacian Operator

```python
laplacian = cv.Laplacian(img,cv.CV_64F)
```

#### 4.Morphological Gradient (Dilaion - erosion)

```python
kerenl = np.ones((5,5),np.uint8)
morph_gradient = cv.morphologyEx(img,cv.MORPH_GRADIENT,kerenl)
morph_close = cv.morphologyEx(img,cv.MORPH_CLOSE,kerenl)
morph_cross = cv.morphologyEx(img,cv.MORPH_CROSS,kerenl)
morph_dilate = cv.morphologyEx(img,cv.MORPH_DILATE,kerenl)
morph_erode = cv.morphologyEx(img,cv.MORPH_ERODE,kerenl)
```

```python
show (
    ["Original","Canny","Sobel","laplacian","morph Gradient","morph Close","morph_cross","morph dilate","morph erode"],
    [img , canny,sobel,laplacian,morph_gradient,morph_close,morph_cross,morph_dilate,morph_erode]
)
```
