#### filtering and blurring

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(img , title=""):
    plt.figure(figsize=(5,5,))
    plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
img = cv.imread("pictures/dog.jpg")
show(img,"Original image")
```

#### Average Blur

```python
avg = cv.blur(img,(7,7,))
show(avg,"Average Blurred")
```

#### median Blur

```python
meadian = cv.medianBlur(img,7)
show(meadian,"Median Blur")
```

#### Gaussian Blur

```python
gaussian = cv.GaussianBlur(img,(7,7),sigmaX=1.5)
show(gaussian,"Gaussian")
```

#### Bilateral Filter

```python
bilateral = cv.bilateralFilter(img, d=19, sigmaColor=75,sigmaSpace=75)
show(bilateral,"Bilateral Fillter")
```

#### Custom Filtering

```python
# Example: Sharpening filter
kernel_sharpen = np.array([[0,-1,0],
                           [-1,5,-1],
                           [0,-1,0]], np.float32)

custom =cv.filter2D(img, -1, kernel_sharpen)
show(custom, "STEP 6: filter2D (Sharpen)")
```
